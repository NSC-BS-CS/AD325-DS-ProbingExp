```python
class TableEntry:
    def __init__(self, key, value):
        self.key = key
        self.value = value
        self.in_table = True

    def get_key(self):
        return self.key

    def get_value(self):
        return self.value

    def set_value(self, value):
        self.value = value

    def is_in_table(self):
        return self.in_table

    def set_to_removed(self):
        self.key = None
        self.value = None
        self.in_table = False

    def is_removed(self):
        return (self.key is None) and (self.value is None) and not self.in_table
```

```python
class HashTableWithCount:
    """
    Base class for all open-addressing hash tables.

    Subclasses implement exactly two methods:
        probe(index, key)  — find the slot to insert/update into
        locate(index, key) — find the slot a key lives in (returns -1 if absent)

    Everything else — add, remove, get_value, contains, clear, resizing —
    is inherited from here unchanged. Do not duplicate these in your subclasses.
    """

    def __init__(self, initial_capacity=101):
        self.number_of_entries = 0
        self.locations_used = 0
        self.probes_needed = 0
        self.table_size = self.get_next_prime(initial_capacity)
        self.hash_table = [None] * self.table_size

    # ── Utility — provided, no changes needed ────────────────────────────────

    def get_next_prime(self, n):
        while True:
            if self.is_prime(n):
                return n
            n += 1

    def is_prime(self, n):
        if n <= 1:   return False
        if n <= 3:   return True
        if n % 2 == 0 or n % 3 == 0: return False
        i = 5
        while i * i <= n:
            if n % i == 0 or n % (i + 2) == 0:
                return False
            i += 6
        return True

    def check_size(self, size):
        if size > 2 * 10000:
            raise ValueError("Size exceeds maximum allowable size")

    def get_hash_index(self, key):
        return hash(key) % self.table_size

    def is_hash_table_too_full(self):
        return self.locations_used > self.table_size * 0.5

    def enlarge_hash_table(self):
        old_table = self.hash_table
        self.table_size = self.get_next_prime(2 * self.table_size)
        self.hash_table = [None] * self.table_size
        self.locations_used = 0
        for entry in old_table:
            if entry is not None and entry.is_in_table():
                index = self.get_hash_index(entry.get_key())
                index = self.probe(index, entry.get_key())
                self.hash_table[index] = entry
                self.locations_used += 1

    def reset_probe_count(self):  self.probes_needed = 0
    def get_probes(self):         return self.probes_needed

    # ── Abstract interface — YOU must implement these in each subclass ────────

    def probe(self, index, key):
        raise NotImplementedError("Subclasses must implement probe()")

    def locate(self, index, key):
        raise NotImplementedError("Subclasses must implement locate()")

    # ── Shared operations — provided, inherited by all subclasses ─────────────
    # Do NOT copy these into your subclasses. They call self.probe() and
    # self.locate(), which will resolve to whichever subclass you instantiate.

    def add(self, key, value):
        if key is None or value is None:
            raise ValueError("Cannot add None to the table.")
        old_value = None
        index = self.get_hash_index(key)
        index = self.probe(index, key)
        if self.hash_table[index] is None or self.hash_table[index].is_removed():
            self.hash_table[index] = TableEntry(key, value)
            self.number_of_entries += 1
            self.locations_used += 1
        else:
            old_value = self.hash_table[index].get_value()
            self.hash_table[index].set_value(value)
        if self.is_hash_table_too_full():
            self.enlarge_hash_table()
        return old_value

    def remove(self, key):
        removed_value = None
        index = self.get_hash_index(key)
        index = self.locate(index, key)
        if index != -1:
            removed_value = self.hash_table[index].get_value()
            self.hash_table[index].set_to_removed()
            self.number_of_entries -= 1
        return removed_value

    def get_value(self, key):
        index = self.get_hash_index(key)
        index = self.locate(index, key)
        return self.hash_table[index].get_value() if index != -1 else None

    def contains(self, key):
        return self.locate(self.get_hash_index(key), key) != -1

    def is_empty(self):   return self.number_of_entries == 0
    def get_size(self):   return self.number_of_entries

    def clear(self):
        for i in range(len(self.hash_table)):
            self.hash_table[i] = None
        self.number_of_entries = 0
        self.locations_used = 0
```
